```python
import tensorflow as tf
import numpy as np

# Load the trained model (make sure the path is correct)
predictionmodel = tf.keras.models.load_model("trained_model.keras")

# Load and preprocess images correctly
input_batches = []
for path in image_paths:
    # 1. Load image with target size (128,128)
    image = tf.keras.preprocessing.image.load_img(path, target_size=(128, 128))

    # 2. Convert to array and ensure 3 channels (RGB)
    input_arr = tf.keras.preprocessing.image.img_to_array(image) / 255.0

    # 3. Verify shape is (128,128,3)
    assert input_arr.shape == (128, 128, 3), f"Invalid shape: {input_arr.shape}"

    input_batches.append(input_arr)

# Stack into batch with shape (N,128,128,3)
final_batch = np.stack(input_batches)

# Predict
predictions = predictionmodel.predict(final_batch)
print("Predictions shape:", predictions.shape)
```

```python
result_indices = np.argmax(predictions, axis=1)
print("Predicted class indices:", result_indices)
```
